# Using Geospatial Data to Predict Climate Change

```python
import xarray as xr

ds = xr.open_dataset("Data/data.nc").sel(expver=1)
ds
```

Can we shift today's data into a new column to create a value that represents the next day's value? I.e. if we take today's temperature, and shift it back one record, then we have a metric that we can target.

```python
ds["Tomorrow"] = ds.skt.shift(-1)
```
